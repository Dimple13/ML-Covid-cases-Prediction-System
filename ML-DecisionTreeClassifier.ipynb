```python
#Importing all the required libraries.
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
#Loading the dataset into variable data and printing it.
data = pd.read_csv('/content/decision_covid.csv')
print(data)
```

```python
#Finding the length and shape of the dataset.
print ("Dataset Length: ", len(data))
print ("Dataset Shape: ", data.shape)
```

```python
#Extracting the input and output from the dataset into variables X and Y.
X = data.values[:, 1:-1]
Y = data.values[:, -1]
Y=Y.astype('int')
```

```python
#Splitting the data into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 100)
```

```python
#Defining the model with gini criteria.The tree will have a maximum depth of 3 and the minimum number of samples required to be at a leaf node is taken as 3.
#Then training the model on training data.
clf_gini = DecisionTreeClassifier(criterion = "gini",random_state = 100,max_depth=3, min_samples_leaf=3)
clf_gini.fit(X_train, y_train)
```

```python
# Predicting the X_test values and storing it in y_pred and displaying it.
y_pred = clf_gini.predict(X_test)
print("Predicted values:")
print(y_pred)
```

```python
#Confusion matrix of the predictions.
confusion_matrix(y_test, y_pred)
```

```python
#Calculating and displaying the accuracy of our model.
accuracy_score(y_test,y_pred)*100
```

```python
#Testing the model on a sample input value
test= clf_gini.predict([[1,1,1,2]])
```
